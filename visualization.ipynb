## 1. Multiple plots in 1 figure

### plt.axes([xlo, ylo, width, height]) - different plots on distinct axes
xlo, ylo, width, height - all in range 0, 1 i.e proprotion of figure size
### plt.subplot(m,n,k) - no need to specify axis dimension for distinct plots

````python
plt.axes([0.05, 0.05, 0.425, 0.9])
plt.plot(year, physical_sciences, color='blue')
plt.axes([0.525, 0.05, 0.425, 0.9])
plt.plot(year, computer_science, color='red')
plt.show()
````

````python
plt.subplot(2,2,1)
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

plt.subplot(2,2,2)
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

plt.subplot(2,2,3)
plt.plot(year, health, color='yellow')
plt.title('Health Professions')

plt.subplot(2,2,4)
plt.plot(year, education, color='green')
plt.title('Education')

plt.tight_layout()
plt.show()
````

## 2. Specifying Axis limits, Legends

### plt.axis((xmin,xmax,ymin,ymax)) - specify limits of axes to be shown
Necessary when we want to zoom in, expand the plot and see a portion of plot or set axis range to include important points (like origin)
##### set limits for axis individually:
plt.xlim([xmin,xmax])
plt.ylim((ymin,ymax))

````python
plt.plot(year, computer_science, color='red')
plt.plot(year, physical_sciences, color='blue')
plt.axis([1990,2010,0,50])
plt.xlim()
plt.show()
````

## 3. Annotations in Plots



### plt.annotate(s,xy,xytext, arrowprops)
To annotate a point in graph using arrow and text

````python
plt.plot(year, health, color='yellow', label="Health Professions")
plt.plot(year, education, color='green', label="Education")
plt.legend(loc='lower right')

xmax = year[np.argmax(health)]
ymax = np.max(health)
plt.annotate(s="Inflection", xy=[xmax,ymax], xytext=[xmax+5,ymax-5],arrowprops={"facecolor":"black"})
plt.show()
````
