# 2D Arrays

## 1. MeshGrid
1. Used for visualizing bivariate functions


### meshgrid(u,v)
- Replicates 1-D arrays u and v along different axes to generate 2-D arrays - X, Y.
- X & Y represent the co-ordinates of generated grid.
- Resulting arrays X and Y have identical shapes.
- Their entries taken pairwise represent co-ordinates of a rectangular grid.

## Visualizing bivariate functions
- Data arising from a bivariate function can be sampled using 2D arrays

## Visualizing bivariate distributions
Inferring distributions from data - fundamental problem in data analysis

### Histograms in 1D
- Choose bins(intervals)
- Count realizations within bins, and then plot  
- Straight line segments are the only possible shape for bins in 1D  
**counts, bins, patches = plt.hist(x, bins=25)**

### hist2d() - Rectangular Binning
**plt.hist2d(x, y, bins=(nx, ny), range=((xmin, xmax), (ymin, ymax))**  
- nx, ny - the number of bins to use in the horizontal and vertical directions  
- range(optional) - rectangular region in which the samples are counted in constructing the 2D histogram
  * xmin, xmax - lower and upper limits for the variables on the x-axis
  * ymin, ymax - lower and upper limits for the variables on the y-axis

````python
# Generate a 2-D histogram
plt.hist2d(hp, mpg, bins=(20,20), range=[(40,235), (8,48)])

# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
````

### Pseudocolor Plot - pcolor()
 * Blocky/pixelated
 * Appropriate for data that is inherently continuous

````python
u = np.linspace(-2,2,63)
v = np.linspace(-1,1,33)

X,Y = np.meshgrid(u,v)
Z = X**2/25 + Y**2/4 # should represent grayscale pixels, high values = lighter colors, low=darker

plt.pcolor(Z)
plt.axis('tight') #to clip any extra whitespace
plt.show()
````

- Observations
    * Plot has many different colored pixels instead of grayscale pixels.
    * The plot has large integer values(determined by matrix dimensiobs) as x and y ticks instead of actual co-ordinates used to compute the matrix
- colormap: A color line that maps numeric values to colors in some continuous way. The set of colors used in the plot is determined by a colormap which smoothly maps values to colors

````python
plt.pcolor(Z, cmap='viridis') #'summer', 'autumn', 'winter' and 'spring'; 'coolwarm', 'magma', 'viridis', 'jet'(D); 'Greens', 'Blues', 'Reds', and 'Purples'
plt.colorbar()
plt.show()
````

###  Plot using meshgrid matrices

````python
plt.pcolor(X,Y,Z, cmap='gray') #Axes determined by meshgrid arrays X & Y
plt.show()
````

### Contour Plots
 - For data that varies continuously
 - Visualizing data without displaying all the available sample values, unlike pcolor that displays entire array
 - Smooth curves along which the function represented by the the data has a constant value

````python
plt.contour(Z,30)
plt.show()
````
