## Image as Data
- Images contain data
- Using matplotlib, we can import image and display it.

```python
import matplotlib.pyplot as plt
data = plt.imread('files/balloon.jpg')
plt.imshow(data)
plt.show()
```

- But the computer cannot see the image. It only sees an array of numbers.
- Color images are stored in 3-dimensional arrays.
- The first two dimensions denote the height and width of an image i.e no. of pixels.
- The last dimension corresponds to Red, Green and Blue pixels present in each pixel.

```python
data.shape
```

```python
data[200, 350]
```

Red channel intensity is high because it has a reddish pixel in that position of balloon

```python
data[100,700]
```

Here blue channel intensity is highest because it is indeed a bluish pixel.

## Modifiying Image Data
- We can change the image by changing the image array data
- Let's set green and blue values of pixels to zero

```python
data[:,:,1] = 0
data[:,:,1] = 0
plt.imshow(data)
```

```python
data.flags
```

```python
data.setflags(write=1)
```

```python
data[:,:,1] = 0
data[:,:,2] = 0
plt.imshow(data)
plt.show()
```

Result contains only information of red channel.  
Now let's set pixels in some positions to have full green intensity but no red, blue intensity

```python
#after re-reading original image
data[200:300,750:850] = [0, 255, 0]
plt.imshow(data)
plt.show()
```
